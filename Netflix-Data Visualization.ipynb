### Data_added column change Dtype object to Datetime

```python
df['date_added']=pd.to_datetime(df["date_added"])
```

### Check canvert or not

```python
df.info()
```

### Add the month name  column

```python
df['month']=df['date_added'].dt.month_name()
df
```

### Add the Year column

```python
df['Year']=df['date_added'].dt.year
df
```

### Add the Day Name column

```python
df['Day']=df['date_added'].dt.day_name()
df
```

### Count total no of movie and Tv Show   

```python
show=df.groupby(['type'])['type'].count().reset_index(name='count')
show=show.set_index('type')
show
```

### check director total movies and TV show

```python
Director=df.groupby(['director','type'])['type'].count().reset_index(name='show count')
Director
```

### Top Directors

```python
top10_directors=Director.query("`show count`>=12")
top10_directors
```

### Check country total movies and show

```python
country=df.groupby(['country','type'])['type'].count().reset_index(name='show count')
country
```

### Top country 


```python
top10_country=country.query("`show count` >= 149")
top10_country
```

### Type of movies and show

```python
content_list=df.groupby(['listed_in','type'])['type'].count().reset_index(name='show_count')
content_list
```

### Top content

```python
top10_content=content_list.query("`show_count` >= 177")
top10_content
```

### Check Ratings

```python
show_rating=df.groupby(['rating','type'])['rating'].count().reset_index(name='ratings_count')
show=show_rating.set_index('rating')
show
```

### Check season and duration  of movies and shows

```python
DSC=df.groupby(['duration','type'])['type'].count().reset_index(name='DSC')
DSC
```

### How much  movies and shows released in year.

```python
release_year=df.groupby(['Year','type'])['type'].count().reset_index(name='Total_release')
release_year
```

###  How much  movies and shows released in month

```python
release_month=df.groupby(['month','type'])['type'].count().reset_index(name='Total_release')
release_month
```

### How much  movies and shows released in  a particular day name.

```python
release_Day=df.groupby(['Day','type'])['type'].count().reset_index(name='Total_release')
release_Day
```

```python
plt.title('overall_top10_content',size=20)
plt.xlabel('listed_in',size=15)
plt.ylabel("show_genre_cout",size=15)
sns.barplot(x=top10_content.reset_index()['listed_in'], y=top10_content.reset_index()['show_count'],hue=top10_content.reset_index()['type']);
plt.xticks(rotation = 65)
plt.show()
```
